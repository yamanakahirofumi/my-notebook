# 緩い方

```python
targetPath = '/work/mokobot'
gitFolder = '.git'
objectsFolder = 'objects'
ignoreFolders = ['pack','info']
for root, dirs, files in os.walk(os.path.join(targetPath,gitFolder,objectsFolder)):
    for ignore in ignoreFolders:
        if ignore in dirs:
            dirs.remove(ignore)
    for file in files:
        with open(os.path.join(root, file),'rb') as f:
            rawdata = zlib.decompress(f.read())
            print(rawdata)
```
